**Ex. 1**

```
# Init

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('https://github.com/ddvika/Data-Science-School-2020/blob/main/hw_intermediate/materials/file_example_XLSX_1000.xlsx?raw=true')
```

```
# 1.1

del df['Unnamed: 0']
```

```
# 1.2

df.sort_values('Age')
```

```
# 1.3

female = df[df['Gender'] == 'Female']
male = df[df['Gender'] == 'Male']
```

```
# 1.4

quantity = list(female.groupby(['Age']).size())
ages = list(female.groupby(['Age'])['Age'].max())

plt.bar(ages, quantity)
plt.xlabel('Age')
plt.ylabel('Quantity, Female')
plt.show()

quantity = list(male.groupby(['Age']).size())
ages = list(male.groupby(['Age'])['Age'].max())

plt.bar(ages, quantity)
plt.xlabel('Age')
plt.ylabel('Quantity, Male')
plt.show()
```

```
# Init

df = pd.read_csv('https://raw.githubusercontent.com/ddvika/Data-Science-School-2020/main/hw_intermediate/materials/appl_1980_2014.csv')
```

```
# 2.1

del df['Unnamed: 0']
```

```
# 2.2

df.dtypes
```

```
# 2.3

df['Date'] = df['Date'].astype('datetime64[ns]')
```

```
# 2.4

df.index = df['Date']
```

```
# 2.5

df.groupby(['Date']).filter(lambda date: len(date) > 1)
```

```
# 2.6

df[(df['Date'].dt.is_month_end) & (df['Date'].dt.dayofweek == 4)]
```

```
# 2.7

df['Date'].max().toordinal() - df['Date'].min().toordinal()
```

```
# 2.8

import numpy as np

(df['Date'].max() - df['Date'].min()) // np.timedelta64(1, 'M')
```

```
# 2.9

adj_close = list(df['Adj Close'])
date = list(df['Date'])

plt.figure(figsize=(13.5,9))
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.plot(date, adj_close)
```

```
# 2.15 Aditional

# Dynamic between Open and Close

date = list(df['Date'])
dyn = list(df['Close'] - df['Open'])

plt.figure(figsize=(20, 10))
plt.grid()
plt.xlabel('Date')
plt.ylabel('Dynamic')         
plt.plot(date, dyn, 'b--')
```
