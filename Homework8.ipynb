Init system of linear equations:

$$
\begin{cases}
x_1 + x_2 - x_3 - 2x_4 = 0, \\
2x_1 + x_2 - x_3 + x_4 = -2, \\
x_1 + x_2 - 3x_3 + x_4 = 4
\end{cases}
$$

Augmented matrix:

$$
\begin{pmatrix}
\left.
\begin{matrix}
1 & 1 & -1 & -2 \\
2 & 1 & -1 & 1 \\
1 & 1 & -3 & 1
\end{matrix} \;
\right |
\begin{matrix}
0 \\ -2 \\ 4
\end{matrix}
\end{pmatrix}
$$

We have 4 variables and only 3 equations. The system is can not definitely solved.

Row 3 - Row 1:

$$
\begin{pmatrix}
\left.
\begin{matrix}
1 & 1 & -1 & -2 \\
2 & 1 & -1 & 1 \\
0 & 0 & -4 & 3
\end{matrix} \;
\right |
\begin{matrix}
0 \\ -2 \\ 4
\end{matrix}
\end{pmatrix} \;
\Rightarrow \;
- 4x_3 + 3 x_4 = 4
\Rightarrow \;
-3x_3 + \frac{3}{4}x_4 - 1 = 0
\Rightarrow \;
x_3 = \frac{3}{2}x_4 - 2
$$

Row 3 - Row 2:

$$
\begin{pmatrix}
\left.
\begin{matrix}
1 & 1 & -1 & -2 \\
2 & 1 & -1 & 1 \\
-1 & 0 & -2 & 0
\end{matrix} \;
\right |
\begin{matrix}
0 \\ -2 \\ 6
\end{matrix}
\end{pmatrix} \;
\Rightarrow \;
-x_1 - 2x_3 = 6 \;
\Rightarrow \;
x_1 = - 2x_3 - 6 \;
\Rightarrow \;
x_1 = - 2 \cdot (\frac{3}{2}x_4 - 2) - 6 \;
\Rightarrow \;
x_1 = - 3x_4 - 2
$$

And
$$
x_2 = 2x_4 + x_3 - x_1 = 2x_4 + \frac{3}{2}x_4 - 2 + 3x_4 + 2 \;
\Rightarrow \;
x_2 = \frac{13}{2}x_4
$$

General solution:

$$
x_1 = -3x_4 - 2, \; x_2 = \frac{13}{2}x_4, \; x_3 = \frac{3}{2}x_4 - 2
$$

The system has an infinitely many sulutions.

Particular solution:

$\sqsupset x_4$ is even $\Rightarrow x_4 = 2n, \forall \; n \in $ **Z**. Then:

$$
x_1 = -6n - 2, \; x_2 = 13n, \; x_3 = 3n - 2, \; x_4 = 2n
$$

**Ex. 2**

$a) \;\;\; f(x) = x \;$ -- Identity,

$
b) \;\;\; f(x) = \left \{
\begin{array}{1}
0, & x < 0 \\
1, & x \geqslant 0
\end{array} \right . \; 
$ -- Binary step,

$c) \;\;\; f(x) = \tanh x = \frac{e^x - e^{-x}}{e^x + e^{-x}} \;$ -- Hyperbolic tangent,

$d) \;\;\; f(x) = \left \{
\begin{array}{1}
0, & x < 0 \\
x, & x \geqslant 0
\end{array} \right . \;
$ -- ReLU,

$e) \;\;\; f(x) = \ln(1 + e^x) \; $ -- Softplus,

$f) \;\;\; f(x) = \sin x \; $ -- Sinusoid.

$a)$

1. Derivative:

$$
\frac{df}{dx} = 1
$$

2. Codomain:

$$
E(y) \in R
$$

3. Monotonicity:

$$
True
$$

```
import numpy as np
import matplotlib.pyplot as plt

def f(x):    
    return x
    
x = np.arange(-5, 5)

plt.plot(x, f(x))
plt.title('Identity', fontsize = 16)
plt.xlabel('x', fontsize = 12)
plt.ylabel('f(x)', fontsize = 12)
plt.grid(True)
plt.show()
```

$b)$

1. Derivative:

$$
\frac{df}{dx} = 0, \; \forall x \in R
$$

2. Codomain:

$$
E(y): y = \{0, 1\}
$$

3. Monotonicity:

$$
False
$$

```
def f(x):
    return [1 if p >= 0 else 0 for p in x]

x = list(np.arange(-5, 5, 0.1))

plt.plot(x, f(x))
plt.title('Binary Step', fontsize = 16)
plt.xlabel('x', fontsize = 12)
plt.ylabel('f(x)', fontsize = 12)
plt.grid(True)
plt.show()
```

$c)$

1. Derivative:

$$
\frac{df}{dx} = \frac{d}{dx}(e^x - e^{-x})(e^x + e^{-x})^{-1} \\
\frac{df}{dx} = (e^x - (-e^{-x}))(e^x + e^{-x})^{-1} \cdot (-1)(e^x - e^{-x})^2(e^x + e^{-x})^{-2} = \frac{e^x + e^{-x}}{e^x + e^{-x}} - \left( \frac{e^x - e^{-x}}{e^x + e^{-x}}\right)^2 = 1 - \left( \frac{e^x - e^{-x}}{e^x + e^{-x}}\right)^2 \\
\frac{df}{dx} = \frac{(e^x + e^{-x})^2 - (e^x - e^{-x})^2}{(e^x + e^{-x})^2} = \frac{2e^xe^{-x} + 2e^xe^{-x}}{(e^x + e^{-x})^2} = \frac{4}{(e^x + e^{-x})^2} = \frac{4}{4 \cosh^2 x} = \frac{1}{\cosh^2 x} = \cosh^{-2}x
$$

2. Codomain:

$$
e^x > 0, \; e^{-x} > 0, \; \forall x \in R \Rightarrow e^x + e^{-x} > 0\\
E(y) \in R
$$

3. Monotonicity:

$$
True
$$

```
from math import exp

def f(x):
    return [(exp(p) - exp(-p)) / (exp(p) + exp(-p)) for p in x]
    
a = list(np.arange(-5, 5, 0.1))

plt.plot(a, f(a))
plt.title('Hyperbolic Tangent', fontsize = 16)
plt.xlabel('x', fontsize = 12)
plt.ylabel('f(x)', fontsize = 12)
plt.grid(True)
plt.show()
```

$d)$

1. Derivative:

$$
\frac{df}{dx} = 
\left \{
\begin{array}{1}
0, & x < 0 \\
1, & x \geqslant 0
\end{array}
\right.
$$

2. Codomain:

$$
E(y) \in \; [0, \infty)
$$

3. Monotonicity:

$$
True
$$

```
def f(x):
    return [p if p >= 0 else 0 for p in x]

x = list(np.arange(-10, 10, 0.1))

plt.plot(x, f(x))
plt.title('ReLU', fontsize = 16)
plt.xlabel('x', fontsize = 12)
plt.ylabel('f(x)', fontsize = 12)
plt.grid(True)
plt.show()
```

$e)$

1. Derivative:

$$
\frac{df}{dx} = \frac{1}{1 + e^x} \cdot e^x = \frac{e^x + 1 - 1}{e^x + 1} = 1 - \frac{1}{e^x + 1}
$$

2. Codomain:

$$
e^x > 0, \; \forall x \in R \Rightarrow 1 + e^x > 0 \\ 
E(y) \in R
$$

3. Monotonicity:

$$
True
$$

```
from math import log, exp

def f(x):    
    return [log(1 + exp(p)) for p in x]
    
x = list(np.arange(-5, 5, 0.1))

plt.plot(x, f(x))
plt.title('Identity', fontsize = 16)
plt.xlabel('x', fontsize = 12)
plt.ylabel('f(x)', fontsize = 12)
plt.grid(True)
plt.show()
```
